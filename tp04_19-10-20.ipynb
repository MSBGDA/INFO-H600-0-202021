```python
def word_in_file(word,fname): 
    f = open(fname) 
    line = f.readline() 
    while line != '' and line.strip() != word: 
        line = f.readline() 
    return line != ''
```

```python
def longest_word(fname): 
    f = open(fname) 
    max_pair = ("", 0) 
    for line in f.readlines(): 
        word = line.strip() 
        if len(word) > max_pair[1]:
            max_pair = (word, len(word)) 
    return max_pair
    
```

```python
def occurrences(mot): 
    d = {} 
    for c in mot: 
        print(c)
        d[c] = d.get(c, 0) + 1
        print(d[c], d)
    return d

print(occurrences("banane"))
```

```python
def occurrences2letters(sequence): 
    dico = {} 
    for i in range(len(sequence)-1): 
        key = sequence[i]+sequence[i+1] 
        dico[key] = dico.get(key,0) + 1 
    return dico
```

```python
occurrences2letters("ACCTAGCCATGTAGAATCGCCTAGGCTTTAGCTAGCTCTAGCTAGCTG")
```

```python
def loadDico(fileName): 
    f = open(fileName) 
    list_words = f.read().split() 
    dico = {} 
    for word in list_words: 
        dico[word] = None 
    return dico
def isInDico(word, dico): 
    return word.upper() in dico
```

```python
path = "C:/Users/maria/OneDrive/Documents/Unif 2021/computing found. data science/tp04/dico.txt"
dico = loadDico(path)
print(isInDico("absinths", dico))
```

```python
path = "C:/Users/maria/OneDrive/Documents/Unif 2021/computing found. data science/tp04/hamlet.txt"
def words(text): 
    import string 
    itab = string.punctuation # punctuation symbols 
    otab = " " * len(string.punctuation) # to replace by space 
    tab = str.maketrans(itab, otab) # translation table 
    text = text.translate(tab) # perform the translation 
    text = text.upper() 
    #print(text.split)
    return text.split()
```

```python
words("If thou hast any sound, or use of voice, Speak to me. If there be any good thing to be done")
```

```python
def occurrenceswords(fileName): 
    f = open(fileName) 
    text = f.read() 
    wordlist = words(text) 
    dico = {} 
    for word in wordlist: 
        dico[word] = dico.get(word, 0) + 1 
    return dico
```

```python
def wordMostUsed(dico): 
    maxi = 0 
    wordMax = "" 
    for word in dico: 
        if dico[word] > maxi: 
            maxi = dico[word] 
            wordMax = word 
    return (wordMax, maxi)
```

```python
dico = occurrenceswords(path) 
print(wordMostUsed(dico))
```
