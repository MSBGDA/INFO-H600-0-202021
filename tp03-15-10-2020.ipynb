```python
def elder(list_name):
    if len(list_name) == 0:
        print('list empty')
    else:
        (eldest, age_eldest) = list_name[0]
        for (name,age) in list_name:
            if age > age_eldest:
                (eldest, age_eldest) = (name, age)
    return eldest            
```

```python
elder([('Thomas', 20), ('Philippe', 24), ('Axel', 22)])
```

```python
def swap(li, i1, i2):
    li[i1], li[i2] = li[i2], li[i1]
    return li
```

```python
def case_swap(string):
    res = ''
    for i in string:
        if i == i.upper():
            res += i.lower()
        else:
            res += i.upper()         
    return res
```

```python
a = 'HelloWorld' 
case_swap(a)
```

```python
from random import randint
def shuffle(li):
    shuffled_list = []
    while len(shuffled_list) != len(li):
        rdm_index = randint(0,len(li)-1)
        rdm_candidate = li[rdm_index]
        if not(rdm_candidate in shuffled_list):
            shuffled_list.append(rdm_candidate)
    return shuffled_list               
```

```python
a = [9, -8, 12, -44, 5, 9, 4, 0, 3, ] 
shuffle(a)
```

```python
b = []
for i in range(100000):
    b.append(i)
shuffle(b)               #take to much time when len is big
```

```python
from random import randint
def shuffle2(li):                                   #problem if same number in list
    li_clone = li[:]
    shuffled_list = []
    while len(shuffled_list) != len(li):
        rdm_index = randint(0,len(li_clone)-1)
        rdm_candidate = li_clone[rdm_index]
        shuffled_list.append(rdm_candidate)
        del li_clone[li_clone.index(rdm_candidate)]    
    return shuffled_list
```

```python
def longest_list(li):
    if len(li) == 0:
        return "empty list"
    else:
        longest = li[0]
        for i in range(1,len(li)):
            if len(li[i]) > len(longest):
                longest = li[i]
        return longest        
```

```python
a = [[11, -2, 35, -46, 7],[2,1],[14, -2, 3, 1]]
longest_list(a)
```

```python
b = []
longest_list(b)
```

```python
def flatten_list(TwoDlist): 
    res = [] 
    for li in TwoDlist: 
        for element in li: 
            res.append(element) 
    return res
```

```python
flatten_list([[11, -2, 35], [-46, 7, -2], [3, 1]])
```
